```python
def decrire_dataframe(df: pd.DataFrame):
    """
    Retourner un dataframe avec des informations sur les colonnes d'un dataframe
    """
    desc = df.columns.to_frame(name="column").set_index('column') #construire le dataframe r√©sultats
    desc['count'] = df.notnull().sum() #count de values non nulls
    desc['unique'] = df.nunique() # nombre de values unique
    desc['dtype'] = df.dtypes # type de donn√©es dans chque colonne
    desc['count_null']=df.isna().sum() #count de valeurs nulls
    desc['null%']=round((df.isna().sum()/df.shape[0]*100),2)
    desc['duplicated']=df.duplicated(keep=False).sum() # nombre de valeurs en doubles
    return desc

#V√©rifier l'unicit√© de la cl√© primaire
#Creation d'une fonction qui teste l'unicit√© d'une colonne  

def cle_primaire(df,col):
    if len(df)==len(df[col].unique()):
        print('La variable', col, 'est une cl√© primaire')
    else: 
        print('La variable',col, 'n\'est pas une cl√© primaire')

def is_zero(cell_value, color='pink'):
    """
    style pour soulign√© une valeur
    example use:
    df.style.applymap(is_zero,subset=["my_col"],color='red')
    """
    return f'background-color: {color};' if cell_value == 0 else None
```

<font color='RoyalBlue'>**A partir de la matrice de correlation:**

-- <font color='RoyalBlue'>***Nous concervons les variables qui sont significativement corr√©l√©es avec l'objectif du projet.***

```python
# Liste des variables √©nerg√©tiques
infos_energie = [
    'YearsENERGYSTARCertified', 'ENERGYSTARScore', 'SiteEUI(kBtu/sf)', 
    'SiteEUIWN(kBtu/sf)', 'SourceEUI(kBtu/sf)', 'SourceEUIWN(kBtu/sf)', 
    'SiteEnergyUse(kBtu)', 'SiteEnergyUseWN(kBtu)', 'SteamUse(kBtu)', 
    'Electricity(kWh)', 'Electricity(kBtu)', 'NaturalGas(therms)', 
    'NaturalGas(kBtu)', 'TotalGHGEmissions', 'GHGEmissionsIntensity'
]

# Variables cibles
var_cibles = ['SiteEnergyUse(kBtu)', 'TotalGHGEmissions']

# Calcul de la matrice de corr√©lation
matrice_corr = df[infos_energie].corr()

# S√©lection des variables √©nerg√©tiques fortement corr√©l√©es avec au moins une des variables cibles
variables_fortement_correlees = {}
for variable in infos_energie:
    for cible in var_cibles:
        if abs(matrice_corr.at[variable, cible]) > 0.8:  # Seuil de corr√©lation fort, par exemple 0.7
            if variable not in variables_fortement_correlees:
                variables_fortement_correlees[variable] = []
            variables_fortement_correlees[variable].append(cible)

# Afficher les variables √©nerg√©tiques fortement corr√©l√©es
for variable, cibles in variables_fortement_correlees.items():
    print(f"La variable {variable} est fortement corr√©l√©e avec {', '.join(cibles)}.")
```

```python
# Paires de variables fortement corr√©l√©es √† examiner
paires_a_examiner = [
    ('Electricity(kWh)', 'Electricity(kBtu)'), 
    ('NaturalGas(therms)', 'NaturalGas(kBtu)'),
    ('SiteEUI(kBtu/sf)', 'SiteEUIWN(kBtu/sf)'),
    ('SiteEUI(kBtu/sf)', 'SourceEUI(kBtu/sf)'),
    ('SiteEUI(kBtu/sf)', 'SourceEUIWN(kBtu/sf)'),
    ('SiteEUIWN(kBtu/sf)', 'SourceEUIWN(kBtu/sf)'),
    ('SourceEUI(kBtu/sf)', 'SourceEUIWN(kBtu/sf)'),
    ('SourceEUI(kBtu/sf)', 'SiteEUIWN(kBtu/sf)'),
    ('SiteEnergyUse(kBtu)', 'Electricity(kWh)'),
]


# D√©finition des variables cibles
var_cibles = ['SiteEnergyUse(kBtu)', 'TotalGHGEmissions']

# Pour chaque paire, on calcule la moyenne des corr√©lations absolues avec les variables cibles
for var_a, var_b in paires_a_examiner:
    # Assurez-vous que les deux variables sont dans le DataFrame
    if var_a in df.columns and var_b in df.columns:
        cor_a = sum(abs(df[var_a].corr(df[cible])) for cible in var_cibles)
        cor_b = sum(abs(df[var_b].corr(df[cible])) for cible in var_cibles)
        
        mean_cor_a = cor_a / len(var_cibles)
        mean_cor_b = cor_b / len(var_cibles)
        
        # Pr√©f√©rer conserver la variable avec "kBtu" dans son nom
        if 'kBtu' in var_a:
            df.drop(columns=[var_b], inplace=True)
            print(f"{var_b} retir√©e, {var_a} conserv√©e")
        elif 'kBtu' in var_b:
            df.drop(columns=[var_a], inplace=True)
            print(f"{var_a} retir√©e, {var_b} conserv√©e")
        elif mean_cor_a > mean_cor_b:
            df.drop(columns=[var_b], inplace=True)
            print(f"{var_b} retir√©e, {var_a} conserv√©e")
        else:
            df.drop(columns=[var_a], inplace=True)
            print(f"{var_a} retir√©e, {var_b} conserv√©e")

# Afficher les colonnes restantes
print("Colonnes restantes apr√®s s√©lection :")
print(df.columns.tolist())
```

- ***Visualisation : Utilisation d'un diagramme en bo√Æte pour voir la distribution des √©missions de GES en fonction du type de b√¢timent.***
****
- ***Pour l'analyse statistique :***
- ***Un test ANOVA est r√©alis√© pour chaque variable afin de tester si les diff√©rences entre les moyennes des √©missions de GES par cat√©gorie sont statistiquement significatives.***

```python
# D√©finition des groupes de variables cat√©gorielles
groupes_variables = {
    'Types et utilisations de b√¢timent': ['BuildingType', 'PrimaryPropertyType'],
    'Localisation': ['Neighborhood']
}

target = 'TotalGHGEmissions'  # La variable num√©rique cible

# Parcourir chaque groupe de variables cat√©gorielles et effectuer des analyses
for nom_groupe, variables in groupes_variables.items():
    print(f"\nAnalyse du groupe: {nom_groupe}")
    for var in variables:
        plt.figure(figsize=(15,8))
        sns.boxplot(x=var, y=target, data=data)
        plt.title(f'√âmissions de GES en fonction de {var}')
        plt.ylabel('√âmissions de GES (eqCO2)')
        plt.xlabel(var)
        plt.xticks(rotation=45)
        plt.show()

        # Pr√©paration des donn√©es pour l'ANOVA
        unique_categories = data[var].dropna().unique()
        group_data = [data[data[var] == cat][target].dropna() for cat in unique_categories]

        # Test ANOVA pour l'analyse statistique
        if len(group_data) > 1:  # ANOVA n√©cessite au moins deux groupes
            anova_result = f_oneway(*group_data)
            print(f"R√©sultat ANOVA pour {var}: Statistique F = {anova_result.statistic}, p-valeur = {anova_result.pvalue}")
            if anova_result.pvalue < 0.05:
                print("Il existe une diff√©rence statistiquement significative entre les groupes.")
            else:
                print("Il n'y a pas de diff√©rence statistiquement significative entre les groupes.")
        else:
            print(f"Pas assez de donn√©es pour effectuer ANOVA sur {var}")
```

- ***Graph1:On observe que certains types de b√¢timents, comme les campus, ont une m√©diane des √©missions plus √©lev√©e, ce qui pourrait indiquer  que certains types de b√¢timents consomment plus d'√©nergie et produisent donc plus de GES.***
